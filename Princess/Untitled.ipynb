```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from sklearn.impute import SimpleImputer

# Load the dataset
df = pd.read_csv('Dataset-Gomez-Et-al.csv')

# Data Inspection
print("Shape of dataset:", df.shape)
display(df.head())
display(df.info())
display(df.describe(include="all"))

# Check column names to ensure correct referencing in the formula
print("Column names in the dataset:")
print(df.columns)  # Display all the columns to check for any discrepancies

# Strip spaces from column names (if any)
df.columns = df.columns.str.strip()

# Check if the required columns exist
required_column = 'Mean HSA'  # This is the column to check
if required_column not in df.columns:
    print(f"Column '{required_column}' is missing from the dataset. Please verify the column name.")
else:
    print(f"Column '{required_column}' exists.")

# Check for missing values
missing_counts = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_summary = pd.DataFrame({
    "Missing Count": missing_counts,
    "Missing Percent": missing_percent
}).sort_values(by="Missing Percent", ascending=False)
display(missing_summary)

# Impute missing values
imputer = SimpleImputer(strategy="most_frequent")
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# Handle outliers using IQR
def detect_outliers(df, columns):
    outlier_summary = {}
    for col in columns:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_summary[col] = len(outliers)
    return pd.DataFrame.from_dict(outlier_summary, orient="index", columns=["Outlier Count"])

numeric_columns = df.select_dtypes(include=np.number).columns
outliers = detect_outliers(df, numeric_columns)
display(outliers)

# Descriptive Statistics
desc_stats = df.describe().T
display(desc_stats)

# Correlation Heatmap
# Filter numeric columns only for correlation analysis
numeric_columns = df.select_dtypes(include=np.number).columns
correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Numeric Variables")
plt.show()

# Hypothesis Testing
# Example: Perform an ANOVA test to see if vaccine hesitancy is affected by health system accessibility
# Use 'Mean HSA' directly without backticks
model = ols('Vaccine_Hesitancy ~ `Mean HSA`', data=df).fit()  # Corrected to use `Mean HSA` in formula with backticks
anova_result = anova_lm(model)
display(anova_result)

# Alternatively, you can use individual HSA columns if preferred:
# model = ols('Vaccine_Hesitancy ~ HSA1 + HSA2 + HSA3 + HSA4 + HSA5 + HSA6 + HSA7', data=df).fit()
# anova_result = anova_lm(model)
# display(anova_result)

# Pearson Correlation between variables
corr_result = df[['Parental_Attitudes', 'Health_System_Accessibility', 'Intervention_Strategy', 'Vaccine_Hesitancy']].corr()
display(corr_result)

# Regression Analysis Example
# Linear regression to predict vaccine hesitancy based on attitudes, accessibility, and intervention strategies
X = df[['Parental_Attitudes', 'Health_System_Accessibility', 'Intervention_Strategy']]
y = df['Vaccine_Hesitancy']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())

# Visualization
# Bar plot of vaccine hesitancy by parental attitudes
plt.figure(figsize=(8, 6))
sns.barplot(x='Parental_Attitudes', y='Vaccine_Hesitancy', data=df, palette='Set2')
plt.title("Vaccine Hesitancy vs Parental Attitudes")
plt.xlabel("Parental Attitudes")
plt.ylabel("Vaccine Hesitancy")
plt.xticks(rotation=45)
plt.show()

# Boxplot of vaccine hesitancy by intervention strategy
plt.figure(figsize=(8, 6))
sns.boxplot(x='Intervention_Strategy', y='Vaccine_Hesitancy', data=df, palette='Set1')
plt.title("Vaccine Hesitancy by Intervention Strategy")
plt.xlabel("Intervention Strategy")
plt.ylabel("Vaccine Hesitancy")
plt.xticks(rotation=45)
plt.show()

# Save cleaned and analyzed data
df_imputed.to_csv('cleaned_data.csv', index=False)
```
